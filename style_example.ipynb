```python
from matplotlib.ticker import AutoMinorLocator
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.reset_orig()

import numpy as np
import pandas as pd
```

```python
import sacredfig
from sacredfig import cm2in, add_labels_to_axes, restyle_axis

plt.style.use(sacredfig.style)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12 * cm2in, 6 * cm2in), dpi=150)

# Panel A
x = np.linspace(0, 5, 501)
y = x * np.sin(x * np.pi)
z = -(x+.5) * np.sin((x+.5) * np.pi)
axes[0].plot(x, y)
axes[0].plot(x, z, alpha=.5)
axes[0].fill_between(x, y, z, where=y>z, facecolor='#eeeeee',)
axes[0].set(xlim=(0, 5), ylim=(-6, 6), xlabel="Time $t$", ylabel="Perturbations ($Y_t$, $Z_t$)")
axes[0].grid(False, which='major', axis='x')

# Panel B
axes[1].plot(y, z, c='k')
axes[1].set(xlim=(-6, 6), ylim=(-6, 6), xlabel="Y-axis perturbation ($Y_t$)", ylabel="Z-axis perturbation ($Z_t$)")

# Ensure each panel is a perfect square
for ax in axes:
    ax.set_box_aspect(1)

# Adjust space between subplots
fig.subplots_adjust(wspace=0.3)

# Add labels a. and b.
add_labels_to_axes(fig, axes)
```

## Boxplot (using matplotlib)

```python
iris = sns.load_dataset("iris")
```

```python
fig, ax = plt.subplots(figsize=(6 * cm2in, 6 * cm2in), dpi=150)
ax.grid(False, which='major', axis='x')

ax.boxplot([iris.sepal_length.values, iris.sepal_width.values],
           labels=['Sepal length', 'Sepal width'])

ax.set_box_aspect(1)
ax.set(xlabel="Attribute", ylabel="Empirical distribution", ylim=(0, 10));
```

## Boxplot using seaborn with an overlay stripplot

```python
from sacredfig.seaborn import boxplot, stripplot
```

```python
fig, ax = plt.subplots(figsize=(6 * cm2in, 6 * cm2in), dpi=150)
boxplot(data=iris, ax=ax, width=0.4, fliersize=0);
stripplot(data=iris, ax=ax)

ax.set(ylim=(0, 8))
ax.set_box_aspect(1)

fig.autofmt_xdate()

sns.despine(fig)
```
